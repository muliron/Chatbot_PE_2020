```python
import pandas as pd
from sklearn import tree

df = pd.read_csv('Dados_Tratados_1.csv', delimiter=",")
df = df.round(0)


coluna_a_ser_avaliada = df[['PRICE/SQFT']]
colunas_a_serem_testadas = df.drop(columns=['PRICE/SQFT']).values

arvore_de_decisao = tree.DecisionTreeClassifier(max_depth=1000)
arvore_de_decisao.fit(colunas_a_serem_testadas, coluna_a_ser_avaliada)

with open('arvore_projeto.dot', 'w') as f:
    f = tree.export_graphviz(arvore_de_decisao,
                             out_file=f,
                             max_depth=1000,
                             impurity=True,
                             feature_names=list(df.drop(columns=['PRICE/SQFT'])),
                             class_names=['Cara', 'Barata'],
                             rounded=True,
                             filled=True)
    
```

```python
import pandas as pd
from sklearn import tree

df = pd.read_csv('teste.csv')

y_df = df[['comestivel']]
x_df = df.drop(columns=['comestivel']).values

d_tree = tree.DecisionTreeClassifier(max_depth=20)
d_tree.fit(x_df, y_df)

with open('aula.dot', 'w') as f:
    f = tree.export_graphviz(d_tree,
                             out_file=f,
                             max_depth=10,
                             impurity=True,
                             feature_names=list(df.drop(['comestivel'], axis=1)),
                             class_names=['False', 'True'],
                             rounded=True,
                             filled=True)
```
